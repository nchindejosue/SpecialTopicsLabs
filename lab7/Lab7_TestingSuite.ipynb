## Part 2: Reliability Metrics Calculation
Based on the test results above, we calculate the key metrics.

```python
# CELL 3: Metrics Calculation
# Note: Replace these values with the output from Cell 2 if they change

total_tests_run = 7
failed_tests = 0  # Assuming all passed based on the fix
total_loc = 160   # Estimated Lines of Code

# 1. Pass Rate
pass_rate = ((total_tests_run - failed_tests) / total_tests_run) * 100

# 2. Defect Density (Defects per KLOC)
# KLOC = 1000 lines of code
kloc = total_loc / 1000
defect_density = failed_tests / kloc

print("--- SOFTWARE RELIABILITY METRICS ---")
print(f"Total Tests:    {total_tests_run}")
print(f"Failures:       {failed_tests}")
print(f"Pass Rate:      {pass_rate:.2f}%")
print(f"Defect Density: {defect_density:.2f} defects/KLOC")

if pass_rate > 95:
    print("✅ STATUS: HIGH RELIABILITY")
else:
    print("⚠️ STATUS: NEEDS IMPROVEMENT")
```
